```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,explained_variance_score


df_train = pd.read_csv('data/ford.csv')
df_train.head(10)
```

```python
#getting rid of column that would cause conflict
df2= df_train.drop(['model','fuelType','transmission'], axis=1)
print(df2)
```

```python
#creating dummies for objects

# dummy1 = df2['transmission']
# dummies1 = pd.get_dummies(dummy1, drop_first=True)
# dummy2 = df2['fuelType']
# dummies2 = pd.get_dummies(dummy1, drop_first=True)

# df3 = pd.concat(df2,dummies1,dummies2)

# dummy_cols = ['transmission','fuelType']
# df_dummies = pd.get_dummies(df2[dummy_cols], dtype=int)
# df3 = pd.concat([df2, df_dummies], axis=1)
df3 = df2.drop(['tax'], axis =1)
```

### Start of training/testing

```python
#spliting for test/train
# colsTodrop = ['transmission_Automatic','transmission_Manual','transmission_Semi-Auto','fuelType_Diesel',
#              'fuelType_Electric','fuelType_Hybrid','fuelType_Other','fuelType_Petrol','price']
X = df3.drop(['price'], axis =1)
y = df3['price']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.3)
```

```python
# X_train.drop(['transmission_Automatic','transmission_Manual','transmission_Semi-Auto','fuelType_Diesel',
#              'fuelType_Electric','fuelType_Hybrid','fuelType_Other','fuelType_Petrol'],axis=1)

```

```python
#create the random forest model
model =  RandomForestRegressor(n_estimators = 100, random_state =42) 
```

```python
param_grid = {'max_depth': [4, 10, 15, None],
              'max_features': [1.0, 'sqrt'],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4]}

```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=2,verbose=2, n_jobs=-1)

#used to grt info for best score..not actually fitted..i think
grid_search.fit(X_train, y_train)
```

```python
model = RandomForestRegressor(**grid_search.best_params_, n_estimators=1000)

#fitting the model
model = model.fit(X_train, y_train)

#checking the score
score = model.score(X_test, y_test)
print("Model score:", score)
```

```python
print("Best hyperparameters:", grid_search.best_params_)

print("Best score:", grid_search.best_score_)
```

```python
#The training r_sq
print("score:%.2f" %model.score(X_train,y_train))
```

#### Prediction on Training Dataset

```python
#Prediction on training dataset
y_trainPred = model.predict(X_train)
y_trainPred[:10]
```

```python
print("mae: %.2f" %mean_absolute_error(y_train,y_trainPred))
```

```python
print("mse: %.2f" %mean_squared_error(y_train,y_trainPred))
```

#### Predictions on Test Data

```python
y_testPred = model.predict(X_test)
y_testPred[:10]
```

```python
print("R_sq:%.2f" %r2_score(y_test,y_testPred))
```

```python
print("mae: %.2f" %mean_absolute_error(y_test,y_testPred))
```

```python
print("mse: %.2f" %mean_squared_error(y_test,y_testPred)) #Probably not the best metrics to use
```
